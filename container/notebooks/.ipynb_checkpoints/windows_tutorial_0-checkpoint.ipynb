# Windows Tutorial for Obtaining and Plotting GES DISC Data

In order to get and plot data from GES DISC on a windows machine, you are going to need a couple of dependencies. I will assume that you are using an anaconda environment.

First, you will need the netcdf4 package. You can download this by running the follownig command in cmd.

> conda install -c anaconda netcdf4=1.2.4

This packaage allows you to work with scientific data in a conveniently packaged array format.

```python
import netCDF4
```

You are also going to need the basemap package to plot the Earth science data. On windows, a stable release of basemap with Python 3.* is not availalbe.  You can, however, download a developers version of basemap using the following command.

> conda install -c conda-forge basemap=1.0.7

```python
from mpl_toolkits.basemap import Basemap
```

Now, let's import all other dependencies.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Navigate to https://daac.gsfc.nasa.gov/ to search for data sets from GES DISC.  You are going to need an account with https://earthdata.nasa.gov.

Let's look at the some data from MERRA2 (https://gmao.gsfc.nasa.gov/reanalysis/MERRA-2/).

```python
dataset = netCDF4.Dataset("MERRA2_100.statM_2d_slv_Nx.198101.nc4")
```

Note that the dataset used is a .nc4 format. 

Let's see what this dataset contains.

```python
for var in dataset.variables:
    print(var)
```

Let's grab lon, lat, and T2MMEAN.

```python
lon = dataset.variables["lon"]
lat = dataset.variables["lat"]
T2_mean = dataset.variables["T2MMEAN"]
```

```python
T2_mean
```

In this case, T2_mean is the mean 2-meter air temperature. We can grab this sort of information about any of the variables.

```python
T2_mean.long_name
```

```python
T2_mean.units
```

```python
T2_mean.shape
```

Now, let's plot it.

```python
lon.shape
```

```python
lat.shape
```

When we plot using basemap, we need the coordinates and the data to have the same numpy array shape.  Meshgrid can take care of this for us.

```python
lons, lats = np.meshgrid(lon,lat)
```

```python
lons.shape == lats.shape
```

```python
lons.shape
```

Let's grab one slice of the T2 data. Each slice represents a time chunk.

```python
T2_mean_0 = T2_mean[:,:][0,:,:]
```

```python
lons.shape == lats.shape == T2_mean_0.shape
```

Now, we are good to plot.

```python
%matplotlib inline
plt.figure(figsize=(10,10))
plt.title("2 Meter Mean Air Temperature (K)")
b = Basemap(projection="ortho", lon_0=-10, lat_0=11, resolution="l")
b.drawcoastlines()
cs = b.pcolor(lons, lats, T2_mean_0, latlon=True)
cbar = b.colorbar(cs, location="bottom", pad="5%")
cbar.set_label("Temp (K)")
```
