## 1. Downloading the Data

We are going to use the Mirador DAAC search engine to find the appropriate data set.

1. Go to the following URL: http://mirador.gsfc.nasa.gov
2. In the keyword box, type: IMERG
3. Click *Search GES-DISC*
3. On the first search result, in the row *View Files:*, click on the "All" link.
4. Click on the first data link (3B-MO.MS.MRG.3IMERG.20170101-S000000-E235959.01.V04A.HDF5)


Now, the data will be downloaded as an HDF5 (Hierarchical Data Format 5) to your local machine. Move it to the current directory where your Jupyter Notebook is located.

## 2. Read the Data into Python

Let's use the h5py package to import the data into Python.

```python
import h5py
```

```python
dataset = h5py.File("3B-MO.MS.MRG.3IMERG.20170101-S000000-E235959.01.V04A.HDF5", "r")
```

We now have an object that contains all of the data from the above dataset. Since this data is stored in hierarchical format (dictionaries of dictionaries), let's find the keys for the root dictionary.

```python
list(dataset.keys())
```

So the root level dictionary has one key: "Grid". 

```python
list(dataset["/Grid"].keys())
```

The "/Grid" directory contains some atmospheric measurements as well as the latitude and longitude for each measurement. 

Here's the hierarchy of this dataset:

Root
<br>
|&nbsp; &nbsp; -- Grid
<br>
|&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; | &nbsp; -- lat
<br>
|&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; | &nbsp; -- precipitation
<br>
|&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; | &nbsp; -- randomError
<br>
| &nbsp;&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; | &nbsp; -- lon
<br>
|&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; | &nbsp; -- gaugeRelativeWeighting
<br>
|&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; | &nbsp; -- probabilityLiquidPrecipitation

Let's learn how to obtain the data from the "Grid" dictionary.

```python
grid = dataset["/Grid"]
```

```python
precipitation = grid["precipitation"]
```

```python
precipitation.shape
```

```python
precipitation.name
```

```python
precipitation.fillvalue
```

We can grab the data for the precipitation object as a numpy array using the following:

```python
precipitation_data = precipitation[:]
```

```python
precipitation_data.shape
```

Okay, let's get the latitude and longitude data now.

```python
lat = grid["lat"][:]
lon = grid["lon"][:]
```

```python
lat
```

```python
lon
```

We can see that the longitude and latitude stretch across the whole world. Latitude goes from -90 to 90 degrees, and longitude goes from -180 to 180 degrees.

## 3. Plotting the Data

We are going to use matplotlib, numpy, and basemap to plot this data onto a world projection.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap, cm
%matplotlib inline
```

The last line in the above box ensures that any plots stay within the Jupyter Notebook.  

### 3.1 Reshape Data

To use basemap, the latitude, longitude and precipitation data must be the same numpy array shape. This is because basemap projects the data onto the the longitude and latitude coordinates.

```python
lon.shape
```

```python
lat.shape
```

```python
precipitation_data.ndim
```

```python
precipitation_data.shape
```

We need to reconfigure the coordinate data to a 2-dimensional grid. We can do so by using the numpy meshgrid method. Specifically, it must the coordinate data must be a 2-dimensional grid with a shape of (3600,1800) so that it matches the shape and dimensions of the precipitation data.

```python
lats, lons = np.meshgrid(lat,lon)
```

```python
# example of np.meshgrid
xs = np.linspace(0,5,6)
ys = np.linspace(9,12,4)
print("xs")
print(xs, xs.shape)
print('\nys')
print(ys, ys.shape)
print('\nxg')
xg, yg = np.meshgrid(xs, ys)
print(xg, xg.shape)
print('\nyg')
print(yg, yg.shape)
```

Now, let's check the dimensions and shapes of our coordinates and precipitation data.

```python
lats.shape == lons.shape == precipitation_data.shape
```

### 3.2 Masking Data

Let's briefly examine the precipitation data. Let's go back to the website that has the download link: <br>
https://mirador.gsfc.nasa.gov/cgi-bin/mirador/granlist.pl?page=1&dataSet=GPM_3IMERGM&version=03&allversion=03&version=04&allversion=04&keyword=IMERG&pointLocation=(-90,-180),(90,180)&location=(-90,-180),(90,180)&searchType=Location&event=&startTime=2014-03-01&endTime=2017-06-15%2023:59:59&search=&CGISESSID=ee6a7fb42fbc7a41e138d53ae9d934b3&nr=58&temporalres=1%20month&prodpg=http://disc.gsfc.nasa.gov/datacollection/GPM_3IMERGM_V04.html&longname=GPM%20IMERG%20Final%20Precipitation%20L3%201%20month%200.1%20degree%20x%200.1%20degree%20V04&granulePresentation=ungrouped

Last time, we clicked on the actual .HDF5 file.  This time, click on OpeNDaP (next to *One Click Download*). You can also access this site using the following URL: https://gpm1.gesdisc.eosdis.nasa.gov/opendap/hyrax/GPM_L3/GPM_3IMERGM.04/2017/3B-MO.MS.MRG.3IMERG.20170101-S000000-E235959.01.V04A.HDF5.html

This website gives us information about the precipitation variable. The units are *mm/hr*. There is also an attribute known as the *fill value*. For a numpy array, if data is missing for a certain measurement, this fill value will be used (in this case, -9999.9004).

```python
precipitation_data.min()
```

```python
precipitation_data.max()
```

Evidently, some of the precipitation data is not recorded. We will create a masked numpy array to *hide* these fill values for plotting the data. Here's how we can do that.

```python
precipitation_data_masked = np.ma.masked_where(precipitation_data < -9000, precipitation_data)
```

```python
precipitation_data
```

```python
precipitation_data_masked
```

Now, we are ready to plot.

### 3.3 Using Basemap

We are going to setup the boundaries for our map.  We will use:

* Minimum Latitude: -60
* Maximum Latitude: 60
* Minimum Longitude: -180
* Maximum Longitude: 180

```python
latcorners = [-60,60]
loncorners = [-180, 180]
```

Let's configure the coloring scheme.

```python
cmap = cm.GMT_drywet
```

We will use a cylinder projection of the world, and produce a contour map with the precipitation data.

```python
plt.figure(figsize=(20,20))
plt.title("Precipitation Measurement from IMERG")

m = Basemap(projection="cyl", llcrnrlat=latcorners[0], urcrnrlat=latcorners[1], llcrnrlon=loncorners[0], urcrnrlon=loncorners[1])
m.drawcoastlines()

plot_data = m.contour(lons, lats, precipitation_data_masked,cmap=cmap, latlon=True)

cbar = m.colorbar(plot_data,location='right',pad="5%")
cbar.set_label('mm/h')

parallels = np.arange(-60.,61,20.)
m.drawparallels(parallels,labels=[True,False,True,False])

meridians = np.arange(-180.,180.,60.)
m.drawmeridians(meridians,labels=[False,False,False,True])

font = {'weight' : 'bold', 'size' : 20}
plt.rc('font', **font)
```

The Basemap API is quite useful: https://matplotlib.org/basemap/api/basemap_api.html.  Let's try plotting a filled contour.

```python
plt.figure(figsize=(20,20))
plt.title("Precipitation Measurement from IMERG")

m = Basemap(projection="cyl", llcrnrlat=latcorners[0], urcrnrlat=latcorners[1], llcrnrlon=loncorners[0], urcrnrlon=loncorners[1])
m.drawcoastlines()


data_range = np.arange(0, np.max(precipitation_data_masked), 0.25)
plot_data = m.contourf(lons, lats, precipitation_data_masked, data_range,cmap=cmap, latlon=True)

cbar = m.colorbar(plot_data,location='right',pad="5%")
cbar.set_label('mm/h')

parallels = np.arange(-60.,61,20.)
m.drawparallels(parallels,labels=[True,False,True,False])

meridians = np.arange(-180.,180.,60.)
m.drawmeridians(meridians,labels=[False,False,False,True])

font = {'weight' : 'bold', 'size' : 20}
plt.rc('font', **font)
```

Let's try plotting a different projection. A list can be found here: https://matplotlib.org/basemap/users/mapsetup.html. In this case, we will use an Orthographic projection.

```python
plt.figure(figsize=(13,13))
plt.title("Precipitation Measurement from IMERG")

m = Basemap(projection="ortho", lat_0=0, lon_0=125)
m.drawcoastlines()


data_range = np.arange(0, np.max(precipitation_data_masked), 0.25)
plot_data = m.contourf(lons, lats, precipitation_data_masked, data_range,cmap=cmap, latlon=True)

cbar = m.colorbar(plot_data,location='right',pad="5%")
cbar.set_label('mm/h')


font = {'weight' : 'bold', 'size' : 12}
plt.rc('font', **font)
```
